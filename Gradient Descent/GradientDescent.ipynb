```python
# Gerekli k√ºt√ºphaneleri import edelim
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import ipywidgets as widgets
from IPython.display import display, clear_output
import warnings
warnings.filterwarnings('ignore')

# Matplotlib ayarlarƒ±
plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

print("K√ºt√ºphaneler ba≈üarƒ±yla y√ºklendi!")
```

```python
# Gradient Descent Algoritmasƒ±
class GradientDescent:
    def __init__(self, learning_rate=0.01, max_iterations=100, tolerance=1e-6):
        self.learning_rate = learning_rate
        self.max_iterations = max_iterations
        self.tolerance = tolerance
        self.history = []
        
    def optimize(self, func, gradient_func, start_point):
        """
        Gradient descent optimizasyonu
        
        Args:
            func: Optimize edilecek fonksiyon
            gradient_func: Fonksiyonun t√ºrevi
            start_point: Ba≈ülangƒ±√ß noktasƒ±
        """
        x = start_point
        self.history = [(x, func(x))]
        
        for i in range(self.max_iterations):
            gradient = gradient_func(x)
            
            # Yeni nokta hesapla
            x_new = x - self.learning_rate * gradient
            
            # Yakƒ±nsama kontrol√º
            if abs(x_new - x) < self.tolerance:
                break
                
            x = x_new
            self.history.append((x, func(x)))
            
        return x, func(x), len(self.history)

print("Gradient Descent sƒ±nƒ±fƒ± tanƒ±mlandƒ±!")
```

```python
# Test Fonksiyonlarƒ± ve T√ºrevleri
class TestFunctions:
    @staticmethod
    def quadratic(x):
        """f(x) = x¬≤"""
        return x**2
    
    @staticmethod
    def quadratic_derivative(x):
        """f'(x) = 2x"""
        return 2*x
    
    @staticmethod
    def shifted_quadratic(x):
        """f(x) = (x-3)¬≤"""
        return (x-3)**2
    
    @staticmethod
    def shifted_quadratic_derivative(x):
        """f'(x) = 2(x-3)"""
        return 2*(x-3)
    
    @staticmethod
    def cubic(x):
        """f(x) = x¬≥ - 3x"""
        return x**3 - 3*x
    
    @staticmethod
    def cubic_derivative(x):
        """f'(x) = 3x¬≤ - 3"""
        return 3*x**2 - 3
    
    @staticmethod
    def inverse(x):
        """f(x) = 1/x (x > 0 i√ßin)"""
        if x <= 0:
            return float('inf')
        return 1/x
    
    @staticmethod
    def inverse_derivative(x):
        """f'(x) = -1/x¬≤"""
        if x <= 0:
            return float('inf')
        return -1/(x**2)
    
    @staticmethod
    def sine_function(x):
        """f(x) = sin(x) + 2"""
        return np.sin(x) + 2
    
    @staticmethod
    def sine_derivative(x):
        """f'(x) = cos(x)"""
        return np.cos(x)
    
    @staticmethod
    def polynomial(x):
        """f(x) = x‚Å¥ - 4x¬≥ + 4x¬≤"""
        return x**4 - 4*x**3 + 4*x**2
    
    @staticmethod
    def polynomial_derivative(x):
        """f'(x) = 4x¬≥ - 12x¬≤ + 8x"""
        return 4*x**3 - 12*x**2 + 8*x

# Fonksiyon se√ßenekleri
FUNCTIONS = {
    'x¬≤': (TestFunctions.quadratic, TestFunctions.quadratic_derivative, (-5, 5)),
    '(x-3)¬≤': (TestFunctions.shifted_quadratic, TestFunctions.shifted_quadratic_derivative, (-2, 8)),
    'x¬≥ - 3x': (TestFunctions.cubic, TestFunctions.cubic_derivative, (-3, 3)),
    '1/x': (TestFunctions.inverse, TestFunctions.inverse_derivative, (0.1, 5)),
    'sin(x) + 2': (TestFunctions.sine_function, TestFunctions.sine_derivative, (-2*np.pi, 2*np.pi)),
    'x‚Å¥ - 4x¬≥ + 4x¬≤': (TestFunctions.polynomial, TestFunctions.polynomial_derivative, (-1, 4))
}

print("Test fonksiyonlarƒ± tanƒ±mlandƒ±!")
```

```python
# G√∂rselle≈ütirme Fonksiyonu
def visualize_gradient_descent(func_name, learning_rate, start_point, max_iterations):
    """
    Gradient descent algoritmasƒ±nƒ± g√∂rselle≈ütir
    """
    # Fonksiyon bilgilerini al
    func, grad_func, x_range = FUNCTIONS[func_name]
    
    # Gradient descent objesi olu≈ütur
    gd = GradientDescent(learning_rate=learning_rate, max_iterations=max_iterations)
    
    try:
        # Optimizasyonu √ßalƒ±≈ütƒ±r
        final_x, final_y, iterations = gd.optimize(func, grad_func, start_point)
        
        # Grafik olu≈ütur
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
        
        # Sol grafik: Fonksiyon ve gradient descent adƒ±mlarƒ±
        x_plot = np.linspace(x_range[0], x_range[1], 1000)
        y_plot = np.array([func(x) for x in x_plot])
        
        ax1.plot(x_plot, y_plot, 'b-', linewidth=2, label=f'f(x) = {func_name}')
        
        # Gradient descent adƒ±mlarƒ±nƒ± √ßiz
        if len(gd.history) > 1:
            xs = [point[0] for point in gd.history]
            ys = [point[1] for point in gd.history]
            
            # Adƒ±mlarƒ± g√∂ster
            ax1.plot(xs, ys, 'ro-', markersize=8, linewidth=2, alpha=0.7, label='GD Adƒ±mlarƒ±')
            
            # Ba≈ülangƒ±√ß ve biti≈ü noktalarƒ±nƒ± vurgula
            ax1.plot(xs[0], ys[0], 'go', markersize=12, label='Ba≈ülangƒ±√ß')
            ax1.plot(xs[-1], ys[-1], 'rs', markersize=12, label='Sonu√ß')
            
            # Adƒ±m numaralarƒ±nƒ± g√∂ster
            for i, (x, y) in enumerate(gd.history[::max(1, len(gd.history)//10)]):
                ax1.annotate(f'{i}', (x, y), xytext=(5, 5), textcoords='offset points', 
                           fontsize=8, alpha=0.7)
        
        ax1.set_xlabel('x')
        ax1.set_ylabel('f(x)')
        ax1.set_title(f'Gradient Descent: {func_name}')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # Saƒü grafik: Optimizasyon s√ºreci
        if len(gd.history) > 1:
            iterations_list = list(range(len(gd.history)))
            function_values = [point[1] for point in gd.history]
            
            ax2.plot(iterations_list, function_values, 'b-o', linewidth=2, markersize=4)
            ax2.set_xlabel('ƒ∞terasyon')
            ax2.set_ylabel('Fonksiyon Deƒüeri')
            ax2.set_title('Optimizasyon S√ºreci')
            ax2.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
        
        # Sonu√ßlarƒ± yazdƒ±r
        print(f"\\nüìä Sonu√ßlar:")
        print(f"  ‚Ä¢ Fonksiyon: {func_name}")
        print(f"  ‚Ä¢ Learning Rate: {learning_rate}")
        print(f"  ‚Ä¢ Ba≈ülangƒ±√ß Noktasƒ±: {start_point:.3f}")
        print(f"  ‚Ä¢ Final Nokta: {final_x:.6f}")
        print(f"  ‚Ä¢ Final Deƒüer: {final_y:.6f}")
        print(f"  ‚Ä¢ ƒ∞terasyon Sayƒ±sƒ±: {iterations}")
        
        if len(gd.history) > 1:
            improvement = gd.history[0][1] - gd.history[-1][1]
            print(f"  ‚Ä¢ ƒ∞yile≈ütirme: {improvement:.6f}")
        
    except Exception as e:
        print(f"‚ùå Hata: {str(e)}")
        print("  ‚Ä¢ Learning rate √ßok b√ºy√ºk olabilir")
        print("  ‚Ä¢ Ba≈ülangƒ±√ß noktasƒ± uygun olmayabilir")

print("G√∂rselle≈ütirme fonksiyonu hazƒ±r!")
```

```python
# ƒ∞nteraktif Widget Interface
def create_interactive_widget():
    """
    ƒ∞nteraktif widget aray√ºz√º olu≈ütur
    """
    # Widget'larƒ± tanƒ±mla
    function_dropdown = widgets.Dropdown(
        options=list(FUNCTIONS.keys()),
        value='x¬≤',
        description='Fonksiyon:',
        style={'description_width': 'initial'},
        layout=widgets.Layout(width='300px')
    )
    
    learning_rate_slider = widgets.FloatLogSlider(
        value=0.1,
        base=10,
        min=-3,  # 0.001
        max=0,   # 1.0
        step=0.1,
        description='Learning Rate:',
        style={'description_width': 'initial'},
        layout=widgets.Layout(width='400px')
    )
    
    start_point_slider = widgets.FloatSlider(
        value=2.0,
        min=-5.0,
        max=5.0,
        step=0.1,
        description='Ba≈ülangƒ±√ß Noktasƒ±:',
        style={'description_width': 'initial'},
        layout=widgets.Layout(width='400px')
    )
    
    max_iterations_slider = widgets.IntSlider(
        value=50,
        min=10,
        max=200,
        step=10,
        description='Max ƒ∞terasyon:',
        style={'description_width': 'initial'},
        layout=widgets.Layout(width='400px')
    )
    
    run_button = widgets.Button(
        description='üöÄ √áalƒ±≈ütƒ±r',
        button_style='success',
        layout=widgets.Layout(width='200px', height='40px')
    )
    
    output_area = widgets.Output()
    
    # Fonksiyon deƒüi≈üince ba≈ülangƒ±√ß noktasƒ± aralƒ±ƒüƒ±nƒ± g√ºncelle
    def update_start_point_range(*args):
        selected_func = function_dropdown.value
        x_min, x_max = FUNCTIONS[selected_func][2]
        start_point_slider.min = x_min
        start_point_slider.max = x_max
        start_point_slider.value = (x_min + x_max) / 2
    
    function_dropdown.observe(update_start_point_range, names='value')
    
    # √áalƒ±≈ütƒ±r butonu fonksiyonu
    def run_optimization(button):
        with output_area:
            clear_output(wait=True)
            visualize_gradient_descent(
                function_dropdown.value,
                learning_rate_slider.value,
                start_point_slider.value,
                max_iterations_slider.value
            )
    
    run_button.on_click(run_optimization)
    
    # Layout olu≈ütur
    controls = widgets.VBox([
        widgets.HTML(value="<h3>üéõÔ∏è Parametreler</h3>"),
        function_dropdown,
        learning_rate_slider,
        start_point_slider,
        max_iterations_slider,
        widgets.HTML(value="<br>"),
        run_button,
        widgets.HTML(value="<hr>")
    ])
    
    # Widget'larƒ± g√∂r√ºnt√ºle
    display(controls, output_area)
    
    # ƒ∞lk √ßalƒ±≈ütƒ±rma
    with output_area:
        visualize_gradient_descent('x¬≤', 0.1, 2.0, 50)

print("ƒ∞nteraktif widget hazƒ±r!")
```

```python
# ƒ∞nteraktif aray√ºz√º ba≈ülat
create_interactive_widget()
```

```python
# √ñrnek 2: Cubic Fonksiyon - Multiple Local Minima
print("üîç √ñrnek 2: x¬≥ - 3x fonksiyonu (multiple local minima)")
visualize_gradient_descent('x¬≥ - 3x', learning_rate=0.1, start_point=2.0, max_iterations=100)
```
